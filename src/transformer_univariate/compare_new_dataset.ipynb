```python
import pandas as pd
```

```python
df_new = pd.read_parquet("../../data/kaggle_dataset/1min/XLM-EUR.parquet").reset_index()
```

```python
df_old = pd.read_csv("../../data/historical/xlmeur.csv")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,10))
plt.plot(df_old.open, label="old")
plt.plot(df_new[(df_new.open_time>df_old.date.min()) & (df_new.open_time<df_old.date.max())].reset_index().open, label="new")
plt.legend()
plt.show()
```
